## Apply

```python
for index, row in df.iterrows():
    if isinstance(row['categories'], list):
        for cat in row['categories']:
            if cat in df.columns:
                df.loc[index, cat] = True

            else:
                df[cat] = False
                df.loc[index, cat] = True
```

```python
def fun(rating): 
  
    if rating<6: 
        return "ehhh"
  
    elif rating>= 6 and rating<8: 
        return "mehhh"
  
    else: 
        return "uhhh"
```
